# Advanced Visualization of Market Mood Index vs NIFTY Index

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'data/MMI_09-Jun-2024.csv'
data = pd.read_csv(file_path)

data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')

data.head()
```

## Pair Plot

```python
data.columns = [col.strip() for col in data.columns]
sns.pairplot(data[['Market Mood Index', 'Nifty Index']])
plt.suptitle('Pair Plot: Market Mood Index vs NIFTY Index', y=1.02)
plt.show()
```

## Joint Plot

```python
sns.jointplot(x='Market Mood Index', y='Nifty Index', data=data, kind='reg', height=8)
plt.suptitle('Joint Plot with Regression Line: Market Mood Index vs NIFTY Index', y=1.02)
plt.show()
```
