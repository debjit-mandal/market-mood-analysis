# Visualization of Market Mood Index vs NIFTY Index

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'data/MMI_09-Jun-2024.csv'
data = pd.read_csv(file_path)
data.columns = [col.strip() for col in data.columns]

data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')

data.head()
```

## Scatter Plots

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Market Mood Index', y='Nifty Index', data=data)
plt.xlabel('Market Mood Index')
plt.ylabel('NIFTY Index')
plt.title('Scatter Plot: Market Mood Index vs NIFTY Index')
plt.grid(True)
plt.show()
```

## Heatmaps

```python
plt.figure(figsize=(10, 6))
correlation = data[['Market Mood Index', 'Nifty Index']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap: Market Mood Index vs NIFTY Index')
plt.show()
```
