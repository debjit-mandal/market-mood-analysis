# Descriptive Analysis of Market Mood Index vs NIFTY Index

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/MMI_09-Jun-2024.csv'
data = pd.read_csv(file_path)
data.columns = [col.strip() for col in data.columns]

data.head()
```

## Summary Statistics

```python
mmi_summary = data['Market Mood Index'].describe()
mmi_summary
```

```python
nifty_summary = data['Nifty Index'].describe()
nifty_summary
```

## Trend Analysis

```python
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')

plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Market Mood Index'], label='Market Mood Index')
plt.plot(data['Date'], data['Nifty Index'], label='NIFTY Index')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('Market Mood Index vs NIFTY Index Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
