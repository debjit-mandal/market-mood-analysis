# Moving Average Analysis of Market Mood Index vs NIFTY Index

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/MMI_09-Jun-2024.csv'
data = pd.read_csv(file_path)
data.columns = [col.strip() for col in data.columns]

data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')

data.head()
```

## Moving Average Calculation

```python
data['MMI_30d_MA'] = data['Market Mood Index'].rolling(window=30).mean()
data['NIFTY_30d_MA'] = data['Nifty Index'].rolling(window=30).mean()

data['MMI_90d_MA'] = data['Market Mood Index'].rolling(window=90).mean()
data['NIFTY_90d_MA'] = data['Nifty Index'].rolling(window=90).mean()
```

## Visualization of Moving Averages

```python
plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Market Mood Index'], label='Market Mood Index')
plt.plot(data['Date'], data['MMI_30d_MA'], label='Market Mood Index 30-day MA')
plt.plot(data['Date'], data['MMI_90d_MA'], label='Market Mood Index 90-day MA')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('Market Mood Index with 30-day and 90-day Moving Averages')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Nifty Index'], label='NIFTY Index')
plt.plot(data['Date'], data['NIFTY_30d_MA'], label='NIFTY Index 30-day MA')
plt.plot(data['Date'], data['NIFTY_90d_MA'], label='NIFTY Index 90-day MA')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('NIFTY Index with 30-day and 90-day Moving Averages')
plt.legend()
plt.grid(True)
plt.show()
```
