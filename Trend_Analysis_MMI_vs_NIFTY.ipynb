# Trend Analysis of Market Mood Index vs NIFTY Index

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = 'data/MMI_09-Jun-2024.csv'
data = pd.read_csv(file_path)
data.columns = [col.strip() for col in data.columns]

data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')

data.head()
```

## Linear Trend Analysis

```python
mmi_trend = np.polyfit(data.index, data['Market Mood Index'], 1)
nifty_trend = np.polyfit(data.index, data['Nifty Index'], 1)

mmi_trend_line = np.polyval(mmi_trend, data.index)
nifty_trend_line = np.polyval(nifty_trend, data.index)

plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Market Mood Index'], label='Market Mood Index')
plt.plot(data['Date'], mmi_trend_line, label='MMI Linear Trend', linestyle='--')
plt.plot(data['Date'], data['Nifty Index'], label='NIFTY Index')
plt.plot(data['Date'], nifty_trend_line, label='NIFTY Linear Trend', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('Linear Trend Analysis of Market Mood Index and NIFTY Index')
plt.legend()
plt.grid(True)
plt.show()
```

## Exponential Trend Analysis

```python
mmi_exp_trend = np.polyfit(data.index, np.log(data['Market Mood Index']), 1)
nifty_exp_trend = np.polyfit(data.index, np.log(data['Nifty Index']), 1)

mmi_exp_trend_line = np.exp(np.polyval(mmi_exp_trend, data.index))
nifty_exp_trend_line = np.exp(np.polyval(nifty_exp_trend, data.index))

plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Market Mood Index'], label='Market Mood Index')
plt.plot(data['Date'], mmi_exp_trend_line, label='MMI Exponential Trend', linestyle='--')
plt.plot(data['Date'], data['Nifty Index'], label='NIFTY Index')
plt.plot(data['Date'], nifty_exp_trend_line, label='NIFTY Exponential Trend', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('Exponential Trend Analysis of Market Mood Index and NIFTY Index')
plt.legend()
plt.grid(True)
plt.show()
```
